```python
import numpy as np
import matplotlib.pyplot as plt

class DFT:
    def __init__(self, x):
        self.x = x
        self.N = len(x)
        self.c = np.zeros(self.N, dtype=np.complex_)

    def compute_DFT(self):
        for k in range(self.N):
            sum = 0
            for m in range(self.N):
                sum += self.x[m] * np.exp(1j * 2 * np.pi * k * m / self.N)
            self.c[k] = 1/self.N * sum

    def plot_DFT(self):
        plt.plot(np.abs(self.c))
        plt.xlabel('k')
        plt.ylabel('Amplitude')
        plt.show()

def main():
    x = np.genfromtxt("test2.txt", delimiter=',')
    dft = DFT(x)
    dft.compute_DFT()
    dft.plot_DFT()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

T = 1
delta_t = 0.001
time_steps = np.arange(0, T, delta_t)
x = np.genfromtxt("test.txt", delimiter=',')
N = len(x)


plt.plot(x)
plt.title('My Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()


def x_approx(t):
    sum = 0
    for k in range(N-1):
        sum += x[k] * np.sin(np.pi * (N-1) * t / T - k * np.pi) / (np.pi * (N-1) * t / T - k * np.pi)
    return sum

x_t = np.array([x_approx(t) for t in time_steps])

plt.plot(time_steps, x_t)
plt.xlabel('Time (t)')
plt.ylabel('Signal (x(t))')
plt.title('Approximation of the Discrete Signal')
plt.show()



```
