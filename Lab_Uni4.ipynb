# Discrete Fourier Transform

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.genfromtxt("test2.txt", delimiter=',')

fig = plt.figure(figsize=(12, 7))
plt.plot(x)
plt.title('Signal')
plt.xlabel('k')
plt.ylabel('Value of signal')
plt.show()

x = np.genfromtxt("test2.txt", delimiter=',')
N = len(x)
T1 = (N - 1) / 5
c = np.zeros(N, dtype=np.complex_)
#print(x)

for k in range(N):
    sum = 0
    for m in range(N):
        sum += x[m] * np.exp(1j * 2 * np.pi * k * m / N)
    c[k] = 1/N * sum

#print(c)

fig = plt.figure(figsize=(12, 7))
plt.plot(np.abs(c))
plt.xlabel('k')
plt.ylabel('Amplitude')
plt.show()
```

# DFT using scipy

```python
from scipy.fftpack import fft
 
x = np.genfromtxt("test2.txt", delimiter=',')
N = len(x)

c = fft(x) / N
#fig = plt.figure(figsize=(12, 7))
plt.plot(np.abs(c))
plt.xlabel('k')
plt.ylabel('Amplitude')
plt.show()
```

# Approximation of Discrete Signal

```python
T = 1
delta_t = 0.001
time_steps = np.arange(0, T, delta_t)
x = np.genfromtxt("test.txt", delimiter=',')
N = len(x)

fig = plt.figure(figsize=(12, 7))
plt.plot(x)
plt.title('Signal')
plt.xlabel('k')
plt.ylabel('Value of signal (x(t))')
#plt.xlim(0, 1)
plt.show()


def x_approx(t):
    sum = 0
    for k in range(N):
        sum += x[k] * np.sin(np.pi * (N - 1) * t / T - k * np.pi) / (np.pi * (N - 1) * t / T - k * np.pi)
    return sum

x_t = np.array([x_approx(t) for t in time_steps])

#fig = plt.figure(figsize=(12, 7))
plt.plot(time_steps, x_t)
plt.xlabel('Time (t)')
plt.ylabel('Signal (x(t))')
plt.title('Approximation of the Discrete Signal')
plt.show()
```
