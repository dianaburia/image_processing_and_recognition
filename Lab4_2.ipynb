```python
import numpy as np
from PIL import Image
import scipy.signal as s
import matplotlib.pyplot as plt

im = Image.open("cameraman.tif")
im_arr = np.array(im)

df_dx = s.convolve2d(im_arr, [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], mode='same')
df_dy = s.convolve2d(im_arr, [[-1, -2, -1], [0, 0, 0], [1, 2, 1]], mode='same')

#V1f(x,y) = sqrt((df(x,y) / dx)^2 + (df(x,y) / dy)^2)
grad = np.sqrt(df_dx**2 + df_dy**2)

plt.imshow(np.uint8(grad), cmap='gray')
plt.show()
```

```python
L = 256

neg_im_arr = L - 1 - grad

plt.imshow(neg_im_arr, cmap='gray')
plt.show()
```

```python
plt.hist(neg_im_arr.flatten(), bins=range(0,L))

plt.show()
```

```python
threshold = 70
binary_im_arr = np.where(neg_im_arr >= threshold, 255, 0)
plt.imshow(binary_im_arr, cmap='gray')
plt.show()
```
